```python
import pandas as pd
# Load your uploaded CSV file
df = pd.read_csv("clean_epitopes.csv")
df.head()
```

```python
# Count how many epitopes are predicted as toxic vs non-toxic
print(df['toxicity'].value_counts())

# Count allergenic vs non-allergenic
print(df['allergenicity'].value_counts())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(df['epitope'], df['antigenicity'], color='teal')
plt.title('Antigenicity Score of Epitopes')
plt.xlabel('Epitope')
plt.ylabel('Antigenicity Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['allergenicity'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    colors=['salmon', 'skyblue'],
    title='Allergenicity Distribution'
)
plt.ylabel('')
plt.show()
```

```python
adjuvant = "GIINTLQKYYCRVRGGRCAVLSCLPKEEQIGKCSTRGRKCCRRK"
linker_adju_epitope = "EAAAK"
linker_epitopes = "GPGPG"

epitopes = [
    "EVDPIGHVY",
    "GLEARGEAL",
    "SSSSTLVEV"
]

# Combine all components
vaccine_sequence = adjuvant + linker_adju_epitope + linker_epitopes.join(epitopes)

# Print the final construct
print("Multi-Epitope Vaccine Construct:\n", vaccine_sequence)
```

```python
with open("multi_epitope_vaccine.fasta", "w") as f:
    f.write(">Multi_Epitope_Vaccine\n")
    f.write(vaccine_sequence)
```
