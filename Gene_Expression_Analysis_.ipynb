```python
import pandas as pd
import io

csv_data = """Gene,Expression_Control,Expression_Treated
BRCA1,5.1,7.5
TP53,3.2,2.1
EGFR,6.3,7.0
MYC,4.2,6.1
CDK2,2.9,3.3
KRAS,5.5,6.2
MTOR,3.8,4.5
VEGF,4.7,6.3
BCL2,3.0,2.7
AKT1,4.1,5.9
"""

df = pd.read_csv(io.StringIO(csv_data))
df.to_csv('gene_expression_dataset.csv', index=False)
print("CSV file 'gene_expression_dataset.csv' created successfully.")
```

**Step 2: Load and Analyze in Google Colab**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("gene_expression_dataset.csv")

# View the data
print(df.head())


```

**Step 3: Visualize the Gene Expression**

```python
# Barplot comparison of Control vs. Treated
plt.figure(figsize=(10, 6))
df.set_index("Gene")[["Expression_Control", "Expression_Treated"]].plot(kind="bar")
plt.title("Gene Expression Levels: Control vs. Treated")
plt.ylabel("Expression Level")
plt.xlabel("Gene")
plt.grid(True)
plt.tight_layout()
plt.show()
```
